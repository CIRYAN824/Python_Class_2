# 02 Download data

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
feature = iris['data']
```

```python
feature_names = iris['feature_names']
```

```python
labels = iris['target']
```

# 03 Pandas

```python
import pandas as pd
```

```python
df = pd.DataFrame(feature, columns = feature_names)
```

```python
df.head(10)
```

```python
df['target'] = labels
```

```python
df.tail()
```

```python
df.head()
```

```python
df.describe()
```

# 04 Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

# 4-1 따로따로 그리기!

```python
df0 = df[df['target']==0]
df1 = df[df['target']==1]
df2 = df[df['target']==2]
```

### (1) 히스토그램

```python
df0['sepal length (cm)']
```

```python
plt.figure(figsize=(6,4))

plt.hist(df0['sepal length (cm)'])

plt.title("Sepal length histogram of iris setosa")

plt.show()
```

### (2) 선그래프

```python
plt.figure(figsize=(6,4))

plt.plot(df0['sepal length (cm)'])

plt.title("Sepal length linegraph of iris setosa", fontsize=14)

plt.show()
```

### (3) 막대 그래프

```python
import numpy as np
```

```python
len(df0)
```

```python
x = np.arange(len(df0))
```

```python
x
```

```python
plt.figure(figsize=(6,4))
plt.bar(x, df0['sepal length (cm)'])
plt.title('Sepal length bar graph of iris setosa', fontsize = 14)
plt.show()
```

### (4) 박스플롯

```python
plt.figure(figsize=(6,4))

fig, ax = plt.subplots()
ax.boxplot(df0['sepal length (cm)'])
plt.title("Sepal length linegraph of iris setosa", fontsize=14)

plt.show()
```

### 4-2 같이 그리기

#### (1) 파이 차트

```python
plt.figure(figsize = (6,4))
ratio = [len(df0), len(df1), len(df2)]
labels = ['setosa', 'versicolor', 'virginica']


plt.pie(ratio, labels = labels, autopct='%.1f%%')


plt.show()
```

### (2) 박스플롯

```python
plt.figure(figsize=(6,4))
fig, ax = plt.subplots()
ax.boxplot([df0['sepal length (cm)'], df1['sepal length (cm)'], df2['sepal length (cm)']])
plt.title("Sepal length boxplot of three types of iris")
plt.show()
```

### (3) 산점도

```python
df.columns
```

```python
plt.figure(figsize=(10,7))
sns.scatterplot(data=df, x='sepal length (cm)', y='sepal width (cm)', hue=df['target'], palette='muted')
plt.title('Sepal lenth and width', fontsize = 17)
plt.show()
```

### (4) 히스토그램

```python
plt.figure(figsize=(6,4))
plt.hist(df0['sepal length (cm)'], alpha =0.5)
plt.hist(df1['sepal length (cm)'], alpha =0.5)
plt.hist(df2['sepal length (cm)'], alpha=0.5)
plt.title("Sepal lenth histogram of iris setosa")

plt.legend()
plt.show()
```
