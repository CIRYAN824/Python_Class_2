```python
import numpy as np
import matplotlib.pyplot as plt

from math import factorial
```

```python
def binomial_distribution(k, n, p):
    term1 = factorial(n) / (factorial(k) * factorial(n-k))
    term2 = term1 * p**k * (1-p)**(n-k)
    return term2
```

```python
x = np.arange(16)
```

```python
[2*k for k in range(16)]
```

```python
bin_dist_1 = [binomial_distribution(k, 15, 0.3) for k in range(16)]
```

```python
plt.bar(x, bin_dist_1, color='lightcoral')
plt.text(12, 0.2, 'n, p = 15, 0.3')
plt.show()
```

```python
plt.subplot(3, 1, 1)
bin_dist_1 = [binomial_distribution(k, 15, 0.3) for k in range(16)]
plt.bar(x, bin_dist_1, color='lightcoral')
plt.ylim(0, 0.3)
plt.text(12, 0.25, 'n, p = 15, 0.3')

plt.subplot(3, 1, 2)
bin_dist_1 = [binomial_distribution(k, 15, 0.4) for k in range(16)]
plt.bar(x, bin_dist_1, color='mediumaquamarine')
plt.ylim(0, 0.3)
plt.text(12, 0.25, 'n, p = 15, 0.3')

plt.subplot(3, 1, 3)
bin_dist_1 = [binomial_distribution(k, 15, 0.5) for k in range(16)]
plt.bar(x, bin_dist_1, color='royalblue')
plt.ylim(0, 0.3)
plt.text(12, 0.25, 'n, p = 15, 0.3')

plt.tight_layout()
plt.show()
```

# 02 Normal Distribution

```python
from scipy.special import erf
```

```python
mu = 0
sigma = 1
```

```python
x = np.linspace(-8, 8, 100)
```

```python
def normal_distribution (x, mu, sigma):
    y = (1/np.sqrt(2 * np.pi * sigma**2)) * np.exp(-(x-mu)**2 / (2*sigma**2))
    return y
```

```python
plt.plot(x, normal_distribution(x, mu, sigma), label='normal distribution N(0,1)')
#plt.xlim(-8, 8)
plt.xlabel('x (random variable)')
plt.ylabel('f(x)')
plt.legend(loc='upper right')
plt.tight_layout
plt.show()
```

```python
plt.subplot(3, 1, 1)
plt.plot(x, normal_distribution(x, mu, sigma), label='normal distribution N(0,1)', color='lightcoral')
plt.xlabel('x (random variable)')
plt.ylabel('f(x)')
plt.legend(loc='upper right')
plt.tight_layout
plt.show()

plt.subplot(3, 1, 2)
plt.plot(x, normal_distribution(x, 1, sigma), label='normal distribution N(1,1)', color='mediumaquamarine')
plt.xlabel('x (random variable)')
plt.ylabel('f(x)')
plt.legend(loc='upper right')
plt.tight_layout
plt.show()

plt.subplot(3, 1, 3)
plt.plot(x, normal_distribution(x, 2, sigma), label='normal distribution N(2,1)', color='royalblue')
plt.xlabel('x (random variable)')
plt.ylabel('f(x)')
plt.legend(loc='upper right')
plt.tight_layout
plt.show()
```

```python
from math import exp
```

```python
def poisson_distribution(n, lambd):
    y = (lambd**n) * exp(-lambd) / factorial(n)
    return y
```

```python
y = [poisson_distribution(n,10) for n in range(40)]
```

```python
x = np.arange(40)
y1 = [poisson_distribution(n,10) for n in range(40)]
plt.bar(x, y1, color = 'lightcoral', alpha=0.5, label ='lambda = 10')

y2 = [poisson_distribution(n,5) for n in range(40)]
plt.bar(x, y2, color='mediumaquamarine', alpha=0.5, label ='lambda = 5')

y3 = [poisson_distribution(n,3) for n in range(40)]
plt.bar(x, y3, color='royalblue', alpha=0.5, label ='lambda = 3')

plt.ylim(0, 0.25)
plt.legend()
```
